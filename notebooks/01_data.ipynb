```python
# ============================================================
# CELL 0 – ENVIRONMENT SETUP (Colab + Local)
# ============================================================
import os, sys

# 1. Detect if running on Colab
IN_COLAB = 'google.colab' in sys.modules

if IN_COLAB:
    PROJECT_ROOT = '/content/hate-speech-detection'
    if not os.path.exists(PROJECT_ROOT):
        os.system('git clone https://github.com/thong7d/hate-speech-detection.git ' + PROJECT_ROOT)
else:
    # Local-first logic: detect root even if running from root folder
    cwd = os.getcwd()
    if os.path.basename(cwd) == 'notebooks':
        PROJECT_ROOT = os.path.abspath(os.path.join(cwd, '..'))
    else:
        PROJECT_ROOT = cwd

# 2. Add src/ to Python path
SRC_PATH = os.path.join(PROJECT_ROOT, 'src')
if SRC_PATH not in sys.path:
    sys.path.insert(0, SRC_PATH)

print(f'✅ Environment ready. PROJECT_ROOT: {PROJECT_ROOT}')
```

```python
# ============================================================
# CELL 1 – EXECUTE DATA ACQUISITION
# ============================================================
import os
from data.download import download_and_extract

print("Initiating Data Acquisition Pipeline...")

# Execute the downloaded script using the config file
config_path = os.path.join(PROJECT_ROOT, "configs", "paths.yaml")
download_and_extract(config_path)

# Verify the final results on Local Storage
raw_vihsd_path = os.path.join(PROJECT_ROOT, "data/raw/vihsd")
print("\n--- Final Verification ---")
if os.path.exists(raw_vihsd_path):
    print(f"Contents of {raw_vihsd_path}:")
    for f in sorted(os.listdir(raw_vihsd_path)):
        print(f" - {f}")
else:
    print("❌ Verification failed. Directory not found.")
```
